```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
csvCh1 = "./F0001CH1.CSV"
csvCh2 = "./F0001CH2.CSV"
```

```python
columnNames = ["Parameters","Value","Empty","X","Y","NaN"]
ch1 = pd.read_csv(csvCh1,names=columnNames)
ch2 = pd.read_csv(csvCh2,names=columnNames)
print(ch1)
```

```python
plt.scatter(ch2.Y,ch1.Y,s=4)
plt.title("Intensity vs Vtune")
plt.xlabel("Intensity")
plt.ylabel("Voltage")
plt.grid()
plt.show()
```

```python
plt.plot(ch2.X,ch2.Y)
plt.title("Vtune vs Time")
plt.xlabel("Time")
plt.ylabel("Vtune")
plt.grid()
```
